```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['skyblue','red','green','purple','orange'],edgecolor='black')
plt.title("Basic bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,36,34,40,41,43,45,50,51,56]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age distribution of customers")
plt.xlabel("Age ranges")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
a=['python','java','c++','ruby']
b=[40,30,20,10]
plt.pie(b, labels=a, colors=['pink', 'blue', 'orange', 'red'], autopct='%1.1f%%',startangle=90)
plt.legend()
plt.title("Pragramming languages")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Simple scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
